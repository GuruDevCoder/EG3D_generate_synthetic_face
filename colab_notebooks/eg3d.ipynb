## 3D reconstruction from single input image
For 3D reconustrution using PTI inversion,we need to follow some steps:
1. Convert pretrained Model (EG3D) to Model parameters
2. Project input image to EG3D latent space
3. After retrieving the latent code (compatible with the generators domain), train the generator 
4. Pretrained Generator (on input image) generates video from multiple viewpoints

### Generate ws - latent code for dataset
For each dataset image, project the image to the generators latent space. Generate the pivot latent code - ws  from the "w_plus" projector and save it. The pretrained EG3D checkpoint weights are required. Specify the number of steps for iterating until find an appropriate projected latent pivot code, corresponding to the input image.

```python
%cd /content/eg3d/eg3d/
for image in os.listdir(dataset_path):
  if '.png' in image:
    image_path = os.path.join(dataset_path, image)
    name = image.split('.')[0]
    embedding_path = os.path.join(out_latent_path, image).split('.')[0]+'.npy'
    cam_path = os.path.join(dataset_path, name + '.npy')
    s = "python run_projector.py --outdir={} --latent_space_type w --num_steps 1000\
    --network={} --sample_mult=2  --image_path={} --c_path={} --embedding_path={} ".format(out_reproduced_path, network,
                                                                                          image_path, cam_path, embedding_path)
    print(s)
    print(os.system(s))
    print("Projection results are stored under", os.path.abspath(output_path)) 
```

### Train Generator and recover pivot latent code

```python
# @title Run PTI inversion on compatible latent code (from w projector)
# This script will automatically read the images in ./eg3d/projector_test_data, and find their pivot latent code in ./eg3d/projector_out
%cd /content/eg3d/eg3d/projector/PTI/
for image in os.listdir(dataset_path):
  if '.png' in image:
    name = image.split('.')[0]
    image_path = os.path.join(dataset_path, image)
    
    s = 'python run_pti_single_image.py --image_path={} --run_name {} \
      --embedding_path {} --out {} --cam_path {} --checkpoint_path {}'.format(image_path, name,
                                                        out_latent_path, out_PTI_path,
                                                        camera_params_path, out_gen_ckp_path)
    print(s)
    print(os.system(s))
    # print("Trained Generator model checkpoint can be found under", os.path.abspath(checkpoint_path))
```

### Reconstruct geometry
The trained checkpoint weights, pivot latent code (embeddings), the input image and camera pose is required. 

```python
for image in os.listdir(dataset_path):
  format = '.obj'
  if '.png' in image:
    name = image.split('.')[0]
    npy_path = os.path.join(out_latent_path, name+".npy")
    s = 'python3 gen_samples_from_given_latent_code.py --image_mode="image" --seeds 2 \
    --outdir={} --trunc=0.7 --npy_path {} \
    --network={} --gen_shapes True --voxel_resolution 256 --shape_format {} \
    --vis_shape False --sample_mult 2'.format(output_path, npy_path, out_gen_ckp_path, format)
    os.system(s)
    print("Results are stored under", output_path)
```

## Generate GT data from EG3D samples
Generating GT data with EG3D is faster and more robust. There is no need to care about the rconstruction error of the input image, like in the PTI inversion method. EG3D is trained on FFHQ and the dataset camera parameters are given for the entire dataset. The only disadvantage is that we use entirely synthetic GT data and hence we have to evaluate the performance of our regressor on real test images. Then we can adjust training with a combination of both types of data.
